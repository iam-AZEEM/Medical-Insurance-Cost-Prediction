### Flask Web Framework

```python
Flask is a lightweight Python web framework used to build web applications, APIs, and 
backend services quickly and efficiently

It is called a micro-framework because it provides only the essentials:
    - URL routing
    - Request handling
    - Response generation
    - Template rendering

Flask is built on top of two important Python libraries:

Werkzeug > Handles HTTP requests/responses and WSGI utilities
Jinja2 > Template engine for rendering HTML

Flask is widely used because it is:
    - Simple and beginner-friendly
    - Flexible
    - Lightweight
    - Easy to scale
    - Great for REST APIs
    - Easy to learn compared to larger frameworks
```

### HTTP Methods in Flask / Web Development

```python
HTTP methods define what action the client wants to perform on a resource.

HTTPS >>  HyperText Transfer Protocol Secure
HTTP  >>  HyperText Transfer Protocol

Methods:
GET	>> Retrieve data
POST >> Create new data
DELETE >> Delete data
PUT >>	Update entire data
PATCH >> Update partial data
```

### Request Object in Flask

```python
The request object in Flask is used to access data sent by the client to the server

request.form >>	Form data sent from HTML forms
request.get_json() or request.json >>	JSON data sent in request body
request.files >> Uploaded files
request.method >>	HTTP method (GET, POST, PUT, DELETE)
request.args >>	Query parameters from URL
```
